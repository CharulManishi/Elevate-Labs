```python
# Univariate Analysis for (analyzing the distribution, central tendency, and spread of data effectively)
```

```python
# Univariate - one variable 
# Univariate Analysis - data analysis that focuses on a single variable
```

```python
data.head()
```

```python
sns.barplot(data = data, x = "class", y = "survived", hue = "sex")
plt.show()
```

```python
sns.lineplot(data = data, x = "class", y = "fare")
plt.show()
```

```python
sns.lineplot(data = data, x = "class", y = "fare", hue = "who")
plt.show()
```

```python
sns.histplot(data = data, x = "age", hue = "who")
plt.show()
```

```python
sns.histplot(data = data, x = "who", hue = "alive", kde = True)
plt.show()
```

```python
sns.countplot(data = data, x = "class", hue = "who")
plt.show()
```

```python
sns.scatterplot(data = data, x = "age", y = "fare", hue = "who")
```

```python
# Pair Plot for showing the distribution of the individual variables
```

```python
sns.pairplot(data, hue = "sex")
plt.figure(figsize=(6, 6))
plt.show()
```

```python
# Violin Plot for examining the relationship
```

```python
sns.violinplot(data = data, x = "age")
plt.show()
```

```python
sns.violinplot(data = data, x = "age", y = "who")
plt.show()
```

```python
# Box Plot for showing the outlier in the data
```

```python
sns.boxplot(data = data, x = "age")
plt.show()
```

```python
sns.boxplot(data = data, x = "age", y = "who")
plt.show()
```

```python
gp = data.groupby("who").agg({"survived" : "mean"})
sns.heatmap(gp)
plt.show()
```

```python
sns.stripplot(data = data, x = "alive", y = "age", hue = "who")
plt.show()
```

```python
sns.swarmplot(data = data, x = "alive", y = "age", hue = "who")
plt.show()
```

```python
sns.catplot(data = data, x = "alive", y = "age", hue = "who")
plt.show()
```

```python
sns.pointplot(data = data, x = "class", y = "age")
plt.show()
```

```python
sns.pointplot(data = data, x = "class", y = "age", hue = "who")
plt.show()
```
